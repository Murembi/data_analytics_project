Organize all the data into a dictionary


```python
#create a dictionery with orginal name of league as key
dict_countries = {
              'Spanish La Liga':'SP1', 'Spanish Segunda Division':'SP2',
              'German Bundesliga':'D1', 
              'English Premier League':'E0', 'English League 1':'E2', 'English League 2':'E3',
             }
```

```python
#get a dictionary element
dict_countries['Spanish La Liga']
```

```python
for league in dict_countries:
    print(dict_countries[league])
```

```python
dict_historical_data = {}
#creating list of league
#leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
# make sure the list name is consistent

for league in dict_countries:
    frames=[]
    for season in range(15, 21):
        file_path = root + str(season) + str(season+1) +'/'+ dict_countries[league]+ '.csv'  # use league variable here
        df = pd.read_csv(file_path, encoding='unicode_escape')
        df.insert(1, 'season', season)
        frames.append(df)   
    df_concat = pd.concat(frames)
    dict_historical_data[league] = df_concat
```
